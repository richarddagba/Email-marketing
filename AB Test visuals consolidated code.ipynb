```
# Import the necessary plotting library
import matplotlib.pyplot as plt

# Define the two groups being compared
groups = ['Personalized', 'Generic']

# Open and Unsubscribe rates for each group
open_rates = [31.2, 19.8]
unsubscribe_rates = [0, 1.1]

# Create X-axis positions for the bars
x = range(len(groups))  # [0, 1]

# Set up the figure size
plt.figure(figsize=(10, 5))

# Plot the Open Rate bars at the default x positions
plt.bar(x, open_rates, width=0.4, label='Open Rate (%)', align='center')

# Plot the Unsubscribe Rate bars slightly to the right of the original bars
plt.bar([p + 0.4 for p in x], unsubscribe_rates, width=0.4, label='Unsubscribe Rate (%)', align='center')

# Set the x-tick labels in between the two sets of bars
plt.xticks([p + 0.2 for p in x], groups)

# Add chart title and Y-axis label
plt.title('Comparison of Engagement Metrics')
plt.ylabel('Percentage (%)')

# Display legend to distinguish the bars
plt.legend()

# Add horizontal gridlines for better readability
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Display the plot
plt.show()
```

```
import matplotlib.pyplot as plt
import numpy as np

# Labels for each metric
labels = ['Open Rate', 'Bounce Rate', 'Unsubscribe Rate']

# Metric values for both email types
personalized = [31.2, 2.1, 0.0]
generic = [19.8, 3.2, 1.1]

# X-axis locations for each group
x = np.arange(len(labels))
width = 0.35  # Width of the bars

# Create the bar chart
fig, ax = plt.subplots()
bars1 = ax.bar(x - width/2, personalized, width, label='Personalized (A)', color='royalblue')
bars2 = ax.bar(x + width/2, generic, width, label='Generic (B)', color='darkorange')

# Add labels and title
ax.set_ylabel('Percentage (%)')
ax.set_title('Key Metric Comparison')
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.legend()

plt.show()
```

```
import seaborn as sns
import matplotlib.pyplot as plt

# 2x2 contingency matrix: rows = email type, columns = open status
# Data: [ [Personalized Opened, Personalized Not Opened], [Generic Opened, Generic Not Opened] ]
data = [[54, 39], [21, 70]]

# Labels for axes
labels = ['Opened', 'Not Opened']
index = ['Personalized', 'Generic']

# Create the heatmap
sns.heatmap(data, annot=True, fmt="d", xticklabels=labels, yticklabels=index, cmap="Blues")
plt.title('Open Rate Contingency Table')
plt.xlabel('Opened Status')
plt.ylabel('Email Type')
plt.show()
```

```
import matplotlib.pyplot as plt

# Email types and engagement counts
labels = ['Personalized', 'Generic']
total_opens = [54, 27]
total_clicks = [0, 0]

# X-axis positions
x = np.arange(len(labels))
width = 0.35

# Plot bars for opens and clicks
fig, ax = plt.subplots()
ax.bar(x - width/2, total_opens, width, label='Total Opens', color='steelblue')
ax.bar(x + width/2, total_clicks, width, label='Total Clicks', color='darkorange')

# Add labels and title
ax.set_ylabel('Count')
ax.set_title('Engagement Metrics Breakdown')
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.legend()

plt.show()
```

```
import pandas as pd
import dash
from dash import html, dcc
import plotly.graph_objects as go

# Data
data = {
    "Group": ["Personalized", "Generic"],
    "Open Rate (%)": [31.20, 19.80],
    "Bounce Rate (%)": [2.10, 3.20],
    "Unsubscribe Rate (%)": [0.00, 1.10]
}
df = pd.DataFrame(data)

# Create the chart
fig = go.Figure()
fig.add_trace(go.Bar(name='Open Rate (%)', x=df['Group'], y=df['Open Rate (%)'], marker_color='teal'))
fig.add_trace(go.Bar(name='Bounce Rate (%)', x=df['Group'], y=df['Bounce Rate (%)'], marker_color='orange'))
fig.add_trace(go.Bar(name='Unsubscribe Rate (%)', x=df['Group'], y=df['Unsubscribe Rate (%)'], marker_color='crimson'))

fig.update_layout(
    title='Engagement Metrics Comparison: Personalized vs Generic Emails',
    xaxis_title='Email Type',
    yaxis_title='Percentage (%)',
    barmode='group',
    template='plotly_white'
)

```
